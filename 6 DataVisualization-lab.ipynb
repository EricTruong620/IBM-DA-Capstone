## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt
QUERY = """
SELECT ConvertedComp
from master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot.hist(bins = 20)
plt.title("Converted Compensation Distribution")
plt.xlabel("Compensation in millions")
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
from master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot.box()
plt.title("Age Distribution")
plt.ylabel("Age Number")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
from master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot.scatter(x = "Age", y = "WorkWeekHrs")
plt.title("Age Vs Hours worked per week")
plt.ylabel("Hours Worked Per week")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
from master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot.scatter(x = "WorkWeekHrs", y = "CodeRevHrs", s = "Age")
plt.title("Work Week Hours vs Code Review Hours")
plt.ylabel("Hours Reviewing Code")
plt.xlabel("Hours Worked Per Week")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, Count(Respondent) as Respondent
from DatabaseDesireNextYear
Group by DatabaseDesireNextYear
order by Respondent Desc
limit 5
"""
df = pd.read_sql_query(QUERY,conn)
labels = df["DatabaseDesireNextYear"]
df.plot.pie(y = "Respondent", labels = labels, autopct = '%1.2f%%')
plt.title("Top 5 Databases Respondents want to learn next year")
plt.figure(figsize=(10, 5))
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
from master
where Age >= 30 and Age <= 35
"""
df = pd.read_sql_query(QUERY,conn)
df.groupby('Age').median().plot.bar(stacked = True)
plt.title("Age 30+ vs Hours Worked")
plt.ylabel("Hours")
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
from master
where Age >= 45 and Age <= 60
"""
df = pd.read_sql_query(QUERY,conn)
df.groupby('Age').median().plot.line()
plt.title("Ages 45+ vs Hours Spent")
plt.ylabel("Hours")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, Count(MainBranch) as count
from master
Group by MainBranch
"""
df = pd.read_sql_query(QUERY,conn)
df.plot.barh(x = "MainBranch")
plt.title("Main Branch: Developers vs Non-Developers")
plt.ylabel("Answers to question")
plt.xlabel("Count")
```
